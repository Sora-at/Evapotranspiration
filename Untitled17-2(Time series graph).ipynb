```python
thailand = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017').filterMetadata('country_co', 'equals', 'TH')
```

```python
# FLDAS Data
FLDAS = ee.ImageCollection('NASA/FLDAS/NOAH01/C/GL/M/V001')
FLDAS = FLDAS.filter(ee.Filter.date('2012-01-01', '2022-06-01'))
```

```python
## FLDAS Surface radiative temperature
ET = FLDAS.select('Evap_tavg')
```
