```python
Map = geemap.Map(center = [14, 100], zoom=5.4)
Map
```

```python
##Importing NO2 data (tropospheric NO2 column number density in mol/m^2) from GEE Image collection
ET = ee.ImageCollection('NASA/FLDAS/NOAH01/C/GL/M/V001').select('Evap_tavg')
```

```python
##Long term analysis (comparison of 2019 and 2020)
ET_2022 = ET.filterDate('2012-03-01', '2012-03-31').mean()#.multiply(1e6) # converted from mol/m^2 into µmol/m^2
#NO2_2020 = NO2.filterDate('2020-01-01', '2020-12-31').mean().multiply(1e6)
```

```python
vizParams = {
  'min': 0.0,
  'max': 0.00005,
  'opacity' : 1.0,
  'palette': [ 'black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']}

Map.addLayer(ET_2022, vizParams, 'ET_2022')

palette = ['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']
Map.add_colorbar(palette,vmin=0.0,vmax=0.00005, layer_name='ET',label='Evapotranspiration [$kg$ $m-2$ $s-1$]', position='bottomright',width=4, format='%.5f')
```

```python
#Setting Vector th
states = ee.FeatureCollection("FAO/GAUL/2015/level1").filter(ee.Filter.eq('ADM0_NAME', 'Thailand'))
Map.addLayer(states, {}, "US States", False)

vis_params = {
    'color': '000000', 
    'pointSize': 3,
    'pointShape': 'circle',
    'width': 1.3,
    'lineType': 'solid',
    'fillColor': '00000000',
}

Map.addLayer(states.style(**vis_params), {}, "Styled vector")
```

```python
#Clip thailand
th = ET_2022.clip(states);
```

```python
vizParams = {
  'min': 0.0,
  'max': 0.00005,
  'opacity' : 1.0,
  'palette': [ 'black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']}

Map.addLayer(th, vizParams, 'ET_2022')
```

```python
#Create map use cartoee
point = ee.Geometry.Point(100.4985, 13.7338)

#th_map = ee.ImageCollection('COPERNICUS/S5P/NRTI/L3_NO2').select('NO2_column_number_density').filterDate('2021-02-01', '2021-5-16').mean()
th_map = th
plt.figure(figsize=(8,10))

region = [97.05, 5.59, 105.73, 20.96]
vizParams = {
  'min': 0.0,
  'max': 0.00005,
  'opacity' : 1.0,
  'palette': ['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']}

#Customizing colorbar
visualization = {'bands':"$Evapotranspiration$ $Average$ [$kg$ $m-2$ $s-1$]", 'min':0.0, 'max':0.00005}
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("",\
        ["black", "blue", "purple", "cyan", "green", "yellow", "red"])

ax1 = cee.get_map(th_map, region=region, vis_params=vizParams)

cee.add_gridlines(ax1,interval=(3,3),linestyle=":")
cee.add_colorbar(ax1,vis_params=visualization,loc='right',cmap=cmap, format='%.5f')

ax1.set_title(label = '$Evapotranspiration$ $Average$ (Jun 2022)',fontsize = 15)


ax1.add_feature(cfeature.LAND)
ax1.add_feature(cfeature.OCEAN)

plt.show()
```
